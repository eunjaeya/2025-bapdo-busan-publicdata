# 각 데이터 중 돼지국밥 정보만 추출

```python
import pandas as pd

df_mg = pd.read_csv('./data/부산맛집 국문 정보.csv')
df_mg = df_mg[df_mg.apply(lambda row: row.astype(str).str.contains('돼지국밥', case=False, na=False).any(), axis=1)]

# 콘텐츠명 컬럼 이름을 가게명 으로 변경

df_mg.rename(columns={'콘텐츠명': '식당명'}, inplace=True)
df_mg = df_mg[['식당명', '주소', '운영 및 시간', '대표메뉴', '이미지URL', '썸네일이미지URL', '상세내용']]
df_mg.head()
```

```python
df_as = pd.read_csv('./data/부산안심식당정보.csv')
df_as = df_as[df_as['업소명'].str.contains('국밥', case=False, na=False)]
df_as[['경도', '위도']] = df_as['좌표'].str.replace('POINT\(|\)', '', regex=True).str.split(' ', expand=True).astype(float)
df_as.rename(columns={'업소명': '식당명'}, inplace=True)
df_as = df_as[['식당명', '주소', '위도', '경도']]
df_as
```

```python
df_tsl = pd.read_csv('./data/택슐랭 선정 식당.csv')
df_tsl = df_tsl[df_tsl.apply(lambda row: row.astype(str).str.contains('돼지국밥', case=False, na=False).any(), axis=1)]
df_tsl = df_tsl[['식당명', '주소', '추천메뉴', '위도', '경도']]
df_tsl
```

```python
df_chgg = pd.read_csv('./data/부산 착한가격업소 조회.csv')
df_chgg = df_chgg[df_chgg.apply(lambda row: row.astype(str).str.contains('돼지국밥', case=False, na=False).any(), axis=1)]
df_chgg.rename(columns={'업소명': '식당명', '지역' : '주소'}, inplace=True)
df_chgg = df_chgg[['식당명', '주소', '소개', '주차여부', '영업시간', '첨부이미지1', '첨부이미지2']]
df_chgg
```

```python
df_mb = pd.read_csv('./data/모범음식점 현황 조회.csv')
df_mb = df_mb[df_mb.apply(lambda row: row.astype(str).str.contains('돼지국밥', case=False, na=False).any(), axis=1)]
df_mb.rename(columns={'업소명': '식당명', '도로명주소' : '주소', '주메뉴' : '대표메뉴'}, inplace=True)
df_mb = df_mb[['식당명', '주소', '대표메뉴', '위도', '경도']]
df_mb
```

```python
df_total_ggg = pd.read_csv('./data/부산광역시 금정구_식품접객업현황_20240925.csv')
df_total_ggg = df_total_ggg[df_total_ggg.apply(lambda row: row.astype(str).str.contains('돼지국밥', case=False, na=False).any(), axis=1)]
df_total_ggg.rename(columns={'업소명': '식당명', '소재지(도로명)' : '주소'}, inplace=True)
df_total_ggg = df_total_ggg[['식당명', '주소']]
df_total_ggg.drop(index=355, inplace= True)
df_total_ggg
```

```python
df_total_bsjg = pd.read_csv('./data/부산광역시_부산진구_일반음식점현황_20241008.csv', encoding='cp949')
df_total_bsjg = df_total_bsjg[df_total_bsjg.apply(lambda row: row.astype(str).str.contains('돼지국밥', case=False, na=False).any(), axis=1)]
df_total_bsjg.rename(columns={'업소명': '식당명', '소재지(도로명)' : '주소'}, inplace=True)
df_total_bsjg = df_total_bsjg[['식당명', '주소']]
df_total_bsjg.dropna(axis=0, inplace= True)
df_total_bsjg
```

# 데이터 병합

```python
df_total_list = [df_total_ggg, df_total_bsjg]
df_total = pd.concat(df_total_list)
df_total.reset_index(inplace = True, drop = True)
df_total.to_csv('./data/전체가게.csv')
```

```python
df_mg.reset_index(inplace=True, drop=True)
df_mg.to_csv('./data/맛집_전처리.csv', index=False)
df_mg.head()
```

```python
df_famous_list = [df_mg, df_as, df_tsl, df_chgg, df_mb]

famous_df = df_mg[['식당명', '주소']]
for temp in df_famous_list[1:]:
    famous_df = pd.concat([famous_df, temp[['식당명', '주소']]])

# 업체명 변경 확인 -> 제외
famous_df = famous_df[famous_df['주소'] != '부산 중구 광복로39번길 30-1  (신창동2가)']

# 시골돼지국밥만 따로 저장
except_df = famous_df[famous_df['식당명'] == '시골돼지국밥']

# 시골돼지국밥을 제외한 나머지 중복 제거
no_except_df = famous_df[famous_df['식당명'] != '시골돼지국밥']
no_except_df = no_except_df.drop_duplicates(subset='식당명', keep='first')
famous_df = pd.concat([no_except_df, except_df], ignore_index=True)

famous_df.reset_index(inplace = True, drop = True)
famous_df.to_csv('./data/유명가게.csv',  index=False)
famous_df.head()
```
